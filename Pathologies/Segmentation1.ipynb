**IMPORTS**

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

**LOAD IMAGE**

```python
img = cv2.imread("AllImages/Desquamation3.jpg")
imgColor = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
imgColorCon = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

**BINARISATION**

```python
retOtsu, thresh1 = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
retOtsu, thresh2 = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
retTriangle, thresh3 = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_TRIANGLE)
print(retOtsu)
print(retTriangle)
```

OTSU: Best value for binarisation
INV: Inversts black and white pixels
Triangle: More efficient method

**DRAW CONTOURS**

```python
contours, hierarchy = cv2.findContours(thresh2, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
# contours = max(contours, key = cv2.contourArea)
imCnt = cv2.drawContours(imgColorCon, contours, -1, (0, 0, 255), 3)
imCntFill =cv2.fillPoly(imCnt, pts =contours, color=(0,100,255))
```

**PLOT**

```python
plt.figure("ImageColor")
plt.imshow(imgColor, cmap="gray")

plt.figure("GrayScale")
plt.imshow(gray, cmap="gray")

plt.figure("retTriangle")
plt.imshow(thresh2, cmap="gray")

plt.figure("ImgContours")
plt.imshow(imCnt, cmap="gray")

plt.show()
```
