```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("AllImages/Desquamation3.jpg")
img_color = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
retOtsu, thresh2 = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
retOtsu, thresh3 = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
retTriangle, thresh4 = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_TRIANGLE)
print(retOtsu)
print(retTriangle)
```

```python
plt.figure("Grayscale")
plt.imshow(gray, cmap="gray")

plt.figure("Otsu")
plt.imshow(thresh2, cmap="gray")

plt.figure("OtsuInv")
plt.imshow(thresh3, cmap="gray")

plt.figure("retTriangle")
plt.imshow(thresh4, cmap="gray")

plt.show()
```
